```python
def temp(x):
    if x == 'Date unknown':
        return x
    elif type(x) == str:
        try:
            result = dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').strftime('%Y-%m')
            return result
        except:
            return x
    elif type(x) == dt.datetime:
        try:
            result = x.strftime('%Y-%m')
            return result
        except:
            return x
    else:
        return "Date unknown"

df1 = pd.read_csv('data/data.csv', index_col=None)
df2 = pd.read_excel('db/album_wEventDate.xlsx', index_col=0)
df2['evdate'] = df2['evdate'].map(lambda x : temp(x))
# df2['evdate'] = df2['evdate'].map(lambda x: x.strftime('%Y-%m'))
df2.head()
```

```python
for i in range(len(df1)):
    temp = df2[df2['albumID'] == df1.loc[i, 'aID']]['evdate']
    if temp.any():
        if temp.values[0] != 'Date unknown':
            df1.loc[i, 'year'] = temp.values[0]
df1.head()
```
